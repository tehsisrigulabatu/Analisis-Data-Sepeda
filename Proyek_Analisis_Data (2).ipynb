## Visualization & Explanatory Analysis

### Pada jam berapa sajakah penyewaan sepeda cenderung meningkat?

```python
# Mengelompokkan data berdasarkan jam
hourly_totals = hour_df.groupby('hr')['cnt'].sum()

# Membuat Line Chart
plt.figure(figsize=(10, 6))
plt.plot(hourly_totals.index, hourly_totals.values, marker='o', linestyle='-', color='b')

plt.title('Total Penyewaan Sepeda per Jam (Semua Hari)', fontsize=14)
plt.xlabel('Jam', fontsize=12)
plt.ylabel('Jumlah Penyewaan', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(alpha=0.3)

plt.show()
```

### Bagaimana perbandingan jumlah penyewaan sepeda pada hari dengan cuaca cerah dengan cuaca mendung (hujan dan bersalju)?

```python
# Mengelompokkan data berdasarkan kategori cuaca
weather_totals = day_df.groupby('weathersit')['cnt'].sum()

# Membuat Bar Chart
plt.figure(figsize=(16, 6))
plt.bar(weather_totals.index, weather_totals.values, color=['skyblue', 'orange', 'gray', 'red'])

# Menambahkan judul dan label
plt.title('Total Penyewaan Sepeda Berdasarkan Cuaca', fontsize=14)
plt.xlabel('Kategori Cuaca', fontsize=12)
plt.ylabel('Total Penyewaan', fontsize=12)
plt.xticks([1, 2, 3], ['Cerah', 'Mendung', 'Hujan Ringan/Salju'])

# Menampilkan chart
plt.show()
```

**Insight:**
- Dari hasil visualisasi data untuk menjawab pertanyaan 1, kita mendapatkan hasil bahwa penyewaan sepeda cenderung meningkat pada pukul 17.00.
- Dari hasil visualisasi data untuk menjawab pertanyaan 2, kita mendapatkan insight bahwa saat cuaca sedang cerah, penyewaan sepeda jauh meningkat (225792 penyewa total) dibanding saat cuaca mendung dan hujan ringan/salju (996858 dan 37869).
